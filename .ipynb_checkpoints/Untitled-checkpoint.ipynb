```python
import numpy as np
from numpy.random import normal, uniform
from scipy.stats import multivariate_normal as mv_norm
import matplotlib.pyplot as plt
```

```python
D = 2
n = 10
for i in range(1, n + 1):
    colormap = plt.cm.gist_ncar
    xs = np.linspace(-10,10,D)
    ys = np.random.multivariate_normal(np.zeros(D), np.eye(D))
    plt.plot(xs,i*ys)
    
plt.show()
plt.grid(True)
plt.title("10 samples of unit 2D gaussian")
plt.xlabel("x")
plt.ylabel("t")
```

```python
D = 100
n = 10
for i in range(1, n + 1):
    colormap = plt.cm.gist_ncar
    xs = np.linspace(-10,10,D)
    ys = np.random.multivariate_normal(np.zeros(D), np.eye(D))
    plt.plot(xs,i*ys)
    
plt.show()
plt.grid(True)
plt.title('10 samples from a %dD gaussian distribution', % n)
plt.xlabel("x")
plt.ylabel("t")
```
